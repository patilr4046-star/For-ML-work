```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df.head()
```

```python
df["Major_Category"] = le.fit_transform(df["Major_Category"])
df["Year_of_Study"] = le.fit_transform(df["Year_of_Study"])
df["Primary_Use_Case"] = le.fit_transform(df["Primary_Use_Case"])
df["Prompt_Engineering_Skill"] = le.fit_transform(df["Prompt_Engineering_Skill"])
```

```python
df = df.astype(int)
```

```python
df.head()
```

```python
X = df.drop("Post_Semester_GPA", axis = 1)
y = df["Post_Semester_GPA"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix , classification_report
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
